```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
row_data=pd.read_csv('1.03. Dummies.csv')
```

```python
row_data.head()
```

```python
data=row_data.copy()
data['Attendance']=data['Attendance'].map({"Yes":1,'No':0})
data.head()
```

```python
data.describe()
```

```python
y=data['GPA']
x1=data[['SAT',"Attendance"]]
```

```python
x=sm.add_constant(x1)
result=sm.OLS(y,x).fit()
result.summary()
```

```python
plt.scatter(data['SAT'],y)
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='red')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='green')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
fig=plt.plot(data['SAT'],yhat_no,lw=2,c='red')
fig=plt.plot(data['SAT'],yhat_yes,lw=2,c='green')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show()
```

```python
plt.scatter(data['SAT'],y,c=data['Attendance'],cmap='RdYlGn_r')
yhat_no=0.6439+0.0014*data['SAT']
yhat_yes=0.8665+0.0014*data['SAT']
yhat=0.0017*data['SAT']+0.275
fig=plt.plot(data['SAT'],yhat_no,c='red',lw=2)
fig=plt.plot(data['SAT'],yhat_yes,c='green',lw=2)
fig=plt.plot(data['SAT'],yhat,c='blue',lw=2)
plt.xlabel('SAT',fontsize=20)
plt.ylabel("GPA",fontsize=20)
plt.show()
```
