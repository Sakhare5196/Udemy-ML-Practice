## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
data=pd.read_csv('iris-dataset.csv')
data.head()
```

```python
data.info()
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
plt.scatter(data['sepal_length'],data['sepal_width'])
plt.xlabel('sepal_length',fontsize=15)
plt.ylabel('sepal_width',fontsize=15)
plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
x=data.iloc[:,0:2]
kmean=KMeans()
kmean.fit(x)
```

```python
cluster_data=data.copy()
idf_cluster=kmean.fit_predict(x)
cluster_data['clustered']=idf_cluster
```

```python
plt.scatter(data['sepal_length'],data['sepal_width'],c=cluster_data['clustered'],cmap='rainbow')
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
from sklearn import preprocessing
```

# Clustering (scaled data)

```python
x_scaled=preprocessing.scale(data)
```

```python
kmean_scaled=KMeans()
kmean_scaled.fit(x_scaled)
```

```python
cluster_scale_data=data.copy()
cluster_scale_data['cluster_scale']=kmean_scaled.fit_predict(x_scaled)
```

```python
plt.scatter(data['sepal_length'],data['sepal_width'],c=cluster_scale_data['cluster_scale'],cmap='rainbow')
```

## Take Advantage of the Elbow Method

### WCSS

```python
wcss=[]
```

### The Elbow Method

```python
for i in range(1,11):
    kmeans= KMeans(i)
    kmeans.fit(x_scaled)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
wcss    
```

```python
number_clusters=range(1,11)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

How many clusters are there?
